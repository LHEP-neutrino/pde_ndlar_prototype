```python
# read the csvs in the pde_files directory and count the number of lines in each
import os
total = 0
num_files = 0
pde_dir = "/global/cfs/cdirs/dune/users/wermelinger/2x2/PDE_study/run1_2x2_JD_muons/"
for filename in os.listdir(pde_dir):
    num_files += 1
    if filename.endswith(".csv"):
        filepath = os.path.join(pde_dir, filename)
        with open(filepath, 'r') as f:
            line_count = sum(1 for line in f)
            total += line_count
        # print(f"{filename}: {line_count} lines")
print(f"Total lines in all CSV files: {total}")
print(f"Number of CSV files: {num_files}")
```
